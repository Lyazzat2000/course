### 1. Знакомство со строками

Изучим такой `тип данных` как `строки`. 

`Строка` – это упорядоченная неизменяемая последовательность символов, заключенных в одинарные или двойные ковычки:
#### "Теперь я знаю, что такое строка"

Ниже будет более подробно объяснено данное определение `строки`.

```python
s = 'Теперь я знаю, что такое строка'
print(s)
```

```python
print(type(s))
```

Строка всегда пишется в ковычках и может содержать любые символы внутри себя, даже числа.

```python
print("1 + 10 =", 1 + 10)
```

Когда мы обособили `"1 + 10"` ковычками , Python вопринял числовое выражение как строку и вывел ее без изменений. Во второй раз мы написали `1 + 10` без ковычек, поэтому Python сложил числа и вывел 11.

Python рассматривает строку как `один объект`, состоящий из нескольких одиночных символов. При этом все символы внутри строки `упорядочены`, а значит у каждого есть свой индекс (начиная с нуля).

Тогда мы можем извлечь символы, ссылаясь на их индексs в строке. Например, извлечем из строки `s` символ под индексом 1:

```python
s = 'Теперь я знаю, что такое строка'
print(s[1])
```

В дополнении у строк, как и у списков, можно извлечь срезы:

```python
s = 'Теперь я знаю, что такое строка'
print(s[15:])
print(s[::-2])
```

Кроме того, строки `неизменяемы`. Это значит, что мы НЕ можем заменить элементы строки на какие-либо другие.

```python
s[2] = 'о'
```

### 2. Операции со строками

Строки можно представить в виде нитки с бусинками, где бусинки – это символы внутри строк. Для Python важно только количество бусинок и их порядок, что написано на бусинках ему безразлично.


С помощью функции `len` мы можем найти количество бусинок в ожерелье.

```python
len(s)
```

В жизни мы можем разобрать ожерелье или объединить несколько нитей в одно ожерелье. Тоже самое можно повторить и со строками.

Представим, что мы хотим объединить 3 нити, чтобы получить 1 длинное ожерелье.

Тогда операция сложения строк будет выглядеть следующим образом:

```python
s1 = 'Я '
s2 = 'изучаю '
s3 = 'Python '
s = s1 + s2 + s3
s
```

Мы можем поменять порядок и получить другое ожерелье в итоге:

```python
s1 = 'Я '
s2 = 'изучаю '
s3 = 'Python '
s_new = s3 + s1 + s2
s_new
```

Но мы не можем поменять порядок символов в первоначальных нитях с бусинками: "Я", "изучаю", "Python"; так как строки `неизменяемы`.

На нить нанизывают только бусинки, то есть только `строки`, а не `числа`. 

```python
s + 555
```

Однако мы можем преобразовать объект из одного типа данных в другой, то есть из обыкновенного шарика тоже сделать бусинку.

```python
shar = str(3.8)
s + shar
```

Так как по определению строки `неизменяемы`, мы не можем за одно действие поменять бусинки на нитке. Но мы можем вытащить необходимые бусинки из старой нити и собрать новую.

```python
s_new = s[:1] + ' с удовольствием ' + s[2:] + '!'
s_new
```

```python
s_new = s[:1] + ' с удовольствием ' + s[2:7] + 'л бы ' + s[9:] + '!'
s_new
```

Кроме операции сложения, существует операция умножения. Уже имеющаяся строка дублируется n количество раз.

```python
s_new = s * 3
print(s_new)
```

### 3. Методы строк

Теперь изучим методы - функции, которые применяютя к объекту в данном случае к строке. Методы строк облегчают работу с ними, они выполняют распространненые операции над строками и делают программный код значительно короче.

Методы `find` и `rfind` находят подстроку в строке, при этом `find` выводит индекс первого вхождения, а `rfind` последнего вхождения.

```python
s = "Лигурийский регулировщик регулировал в Лигурии"
```

Можно представить, что функция `find` читает строку слева направо и ищет слово "Лигур", а функция `rfind` читает строку справа налево и тоже ище слово "Лигур". Когда слово найдено, функция выводит местоположение слова "Лигур" в строке.

```python
s.find("Лигур")
```

```python
s.rfind("Лигур")
```
